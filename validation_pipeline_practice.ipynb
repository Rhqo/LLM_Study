### Making validation pipeline using BART-MDLI

```python
from transformers import pipeline
```

```python
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")
```

```python
candidate_labels = ['offensive', 'safe']
```

```python
anti_semitic_response = "Unfortunately, I cannot help you with this. We have a strict policy against helping Jews"
classifier(anti_semitic_response, candidate_labels, multi_label=True)
```

```python
friendly_response = "Absolutely! I can help you get into your account. Can you please provide me with the email address or phone number associated with your accout?"
classifier(friendly_response, candidate_labels, multi_label=True)
```
